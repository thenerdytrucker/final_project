# Demo Walkthrough



This notebook is a short walkthrough for the final project demo.



It demonstrates:

- a user asking a diabetes-risk question in natural language

- the system parsing the question into model features

- the model generating a prediction response

- an edge case where the input is incomplete and the system reports what is missing



Run the cells from top to bottom during the walkthrough.


```python
from src.train import train_with_config
from src.app import DirectPredictRequest, parse_natural_language_input, predict_direct
from pathlib import Path

import os

import sys


current = Path.cwd().resolve()

project_root = current.parent if current.name == "notebooks" else current

os.chdir(project_root)


if str(project_root) not in sys.path:

    sys.path.insert(0, str(project_root))


config_path = project_root / "configs" / "config.yaml"

metrics = train_with_config(str(config_path))


def run_demo_query(user_question: str):

    parsed = parse_natural_language_input(user_question)

    print("User question:", user_question)

    print("Parsed input:", parsed)

    if not parsed["is_complete"]:

        print("System response: I still need these fields:",
              parsed["missing_features"])

        return None

    prediction = predict_direct(DirectPredictRequest(**parsed["features"]))

    print("Model response:", prediction)

    return prediction


print("Project root:", project_root)

print("Training metrics:", metrics)
```

## Demo 1: Natural-Language Question



This example shows a user asking the system a complete question in natural language. The system extracts the values, runs the model, and returns a response.


```python
demo_question = "My glucose is 148, blood pressure is 72, BMI is 33.6, and age is 50. What is my diabetes risk?"

demo_prediction = run_demo_query(demo_question)
```

## Demo 2: Edge Case



This example shows an incomplete question. The parser extracts what it can and the system reports which fields are still missing instead of trying to run the model with bad input.


```python
edge_case_question = "My glucose is 135 and BMI is 31. Can you tell me my risk?"

edge_case_result = run_demo_query(edge_case_question)
```
